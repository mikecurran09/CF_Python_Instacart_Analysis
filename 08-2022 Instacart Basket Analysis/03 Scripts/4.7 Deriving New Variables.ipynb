## Set-Up

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# Import data sets

df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# Creating subset - first one million rows

df = df_ords_prods_merged[:1000000]
```

# User-Defined Functions

```python
# Define a function - Price Label

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the fuction

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts()
```

# loc( ) Functions

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts()
```

```python
# loc() function is faster - apply to whole df

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range products'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

# For-Loops

```python
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# TASK

```python
# 2 - create a new column 'busiest days'
result2 = []

for value in df_ords_prods_merged['orders_day_of_week']:
  if value in [0,1]:
    result2.append("Busiest days")
  elif value in [3,4]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_days'] = result2
```

```python
# 3 - Check the values of this new column for accuracy

df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

#### Values are accurate as they add up to our frequency check on 'orders_day_of_week'

```python
# 4 - Create a new column containing these labels called “busiest_period_of_day.”

df_hours = df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```
