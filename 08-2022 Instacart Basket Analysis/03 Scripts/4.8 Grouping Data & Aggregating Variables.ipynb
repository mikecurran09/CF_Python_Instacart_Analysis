```python
# 5. Check basis statistics of product prices for each loyalty category
df_ords_prods_deriv.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
# 6. Create a spending flag for each user based on the average price across all their orders

# A. Transform

df_ords_prods_deriv['avg_price'] = df_ords_prods_deriv.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_deriv.head(15)
```

```python
# B. Assign flags

df_ords_prods_deriv.loc[df_ords_prods_deriv['avg_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
df_ords_prods_deriv.loc[df_ords_prods_deriv['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
df_ords_prods_deriv['spender_flag'].value_counts(dropna = False)
```

```python
# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

# A. Transform

df_ords_prods_deriv['median_order_freq'] = df_ords_prods_deriv.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_deriv.head(15)
```

```python
# B. Assign flags

df_ords_prods_deriv.loc[df_ords_prods_deriv['median_order_freq'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_deriv.loc[(df_ords_prods_deriv['median_order_freq'] > 10) & (df_ords_prods_deriv['median_order_freq'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_deriv.loc[df_ords_prods_deriv['median_order_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_deriv['frequency_flag'].value_counts(dropna = False)
```
